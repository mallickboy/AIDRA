```python
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
#extract data from pdf
def load_pdf_data(path):
  loader = DirectoryLoader(path, glob="*.pdf", loader_cls=PyPDFLoader)
  documents = loader.load()
  return documents
```

```python
curr_dir= os.getcwd()
```

```python
corpus_dir= os.path.join(curr_dir,'..','book')
```

```python
extracted_text = load_pdf_data(path=corpus_dir)
```

```python
len(extracted_text)
```

```python
print(extracted_text[4])
```

```python
#split text into chunks
def text_splitter(extracted_data):
  text_splitter = RecursiveCharacterTextSplitter(
      chunk_size = 500,
      chunk_overlap  = 50,
      length_function = len,
  )
  text_chunk= text_splitter.split_documents(extracted_data)
  return text_chunk
```

```python
text_chunk = text_splitter(extracted_text)
```

```python
# checking if the name is present or not
def checkVectorDB(name,PCapi):
    pc = Pinecone(api_key=PCapi)
    existing_indexes = pc.list_indexes()
    for i in range(len(existing_indexes)):
        if existing_indexes[0]['name'] == name:
            return True
    else:
        return False
    
```

```python
name= 'maindb'
checkVectorDB(name,PCapi)
```

```python
from langchain.chains import RetrievalQA
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
# from google.colab import userdata
gemapi= os.getenv('GemAPI')
# GemAPI = userdata.get('GemAPI')
os.environ["GOOGLE_API_KEY"] = gemapi
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash",  # or "gemini-1.5-flash"
    temperature=0.7,
    max_tokens=1024
)
```

```python
from langchain.prompts import PromptTemplate
custom_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are a highly knowledgeable and trustworthy medical expert, specializing in holistic, alternative, and modern medicine. You have access to some of the world’s most authoritative clinical resources, including the Oxford Handbook of Clinical Medicine, Harrison’s Principles of Internal Medicine, The Merck Manual, and other standard medical encyclopedias.

Users will describe their symptoms, ask questions about diseases, or seek treatment advice. You will be provided with relevant excerpts from these medical sources.

Your job is to offer clear, safe, and well-grounded advice — including:
- Accurate diagnosis support (if applicable)
- Natural remedies and holistic practices
- Medications (only if explicitly stated in the provided sources)
- Lifestyle and dietary recommendations

Always act like a professional and empathetic doctor. Never invent or hallucinate treatments. Only make suggestions based on the context given to you from trusted medical literature.

⚠️ Be creative in your communication style, but strictly grounded in the provided medical context.
”

Context:
{context}

Question:
{question}

Answer:
"""
)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriver,
    return_source_documents=True,
    chain_type_kwargs={"prompt": custom_prompt}
)
```

```python
query = "what is Atherosclerosis why it happens"
result = qa_chain.invoke(query)

print("Answer:", result['result'])
# print("\nSource Documents:")
# for doc in result['source_documents']:
    # print(doc.metadata.get('source', 'Unknown Source'))
```
